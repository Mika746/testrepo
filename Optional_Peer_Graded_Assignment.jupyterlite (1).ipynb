```python
#1
#Null hypothesis: The median value of houses is the same whether they are bounded by the Charles River or not
#Alternative hypothesis: The value of houses is the different depends on the river
scipy.stats.ttest_ind(boston_df[boston_df['CHAS']==0]['MEDV'],boston_df[boston_df['CHAS']==1]['MEDV'])
```

```python
#p-value=7.390623170519905e-05, so we reject the null hypothesis. 
#So,we can conclude there is a significant difference in median value of houses bounded by the Charles river.
```

```python
#2
#Null hypothesis: There is no difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)
#Alternative hypothesis: There is difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)
new = boston_df[boston_df['age_group'] == '35 years and younger']['MEDV']
middle = boston_df[boston_df['age_group'] == 'between 35 and 70 years']['MEDV']
old = boston_df[boston_df['age_group'] == '70 years and older']['MEDV']
f_statistic, p_value = scipy.stats.f_oneway(new, middle, old)
print("F_Statistic: {0}, P-Value: {1}".format(f_statistic, p_value))
```

```python
#P-value=1.7105011022702984e-15<<<<0.05
#So, we reject the null hypothesis, so, There is difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)
```
